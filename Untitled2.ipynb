```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

import Configuration 
import Utils
```

```python
params = Configuration.Parameters()
file_locations=params.file_locations
param = params.param
```

```python
data_folder=[params.baseline_locations["Binary_All"][0]]
docs, vocab = Utils.load(data_folder,clean_string=True,remove_stop_words=False,split_for_cv=True)
max_l= Utils.get_dataset_details(docs,vocab)
labels = [sent["y"] for sent in docs]
data = [ sent["text"] for sent in docs]
```

```python
tf_vectorizer = CountVectorizer(max_df=0.5, min_df=1,lowercase = True,
                                    stop_words='english')
dtm_tfidf = tf_vectorizer.fit_transform(data)

lda_tfidf = LatentDirichletAllocation(n_topics=10, random_state=0,learning_method='online')
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tf_vectorizer)
```
